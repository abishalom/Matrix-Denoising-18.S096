```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
```

```python
def solve_ls(X, y):
    inv = np.linalg.inv(np.dot(X.T, X))
    a = np.dot(inv, X.T)
    return np.dot(a, y)

def solve_rr(X, y, lam):
    inv = np.linalg.inv(np.dot(X.T, X) + lam * np.identity(np.shape(X)[1]))
    a = np.dot(inv, X.T)
    return np.dot(a, y)
```

```python
#John's ALS code
def ALS(data_train, data_test, k=2, lam=0.02, max_iter=100):
    '''
    data_train and data_test are lists of tuples (row, column, MSE)
    these need to be integers so we need to map each dataset and lambda value to an integer (MSE need not be integer)
    
    k is rank
    lam is the regularizer on the least squares call in each step
    
    my code also has offsets b_u and b_v
    the prediction for entry (a, i) is (b_u[a] + b_v[i] + np.dot(u[a].T, v[i])[0][0])
    '''
    # size of the problem
    n = max(d[0] for d in data_train)+1 # datasets
    m = max(d[1] for d in data_train)+1 # lambda values
    # which entries are set in each row and column and the MSE
    us_from_v = [[] for i in range(m)]  # II (i-index-set)
    vs_from_u = [[] for a in range(n)]  # AI (a-index set)
    for (a, i, r) in data_train:
        us_from_v[i].append((a, r))
        vs_from_u[a].append((i, r))
    # Initial guesses for u, b_u, v, b_v
    # Note that u and v are lists of column vectors (rows of U, V).
    u, b_u, v, b_v = ([np.random.normal(1/k, size=(k,1)) for a in range(n)],
          np.zeros(n),
          [np.random.normal(1/k, size=(k,1)) for i in range(m)],
          np.zeros(m))
    for itr in range(max_iter):
        if itr%5 == 0:
            print(itr)
        for i in range(len(u)): #run ls on u
            X_mat = np.array([np.append(np.array([1]), v[a[0]].T[0]) for a in vs_from_u[i]])
            y_vec = np.array([[a[1] - b_v[a[0]]] for a in vs_from_u[i]])
            if len(X_mat) > 0:
                sol = solve_ls(X_mat, y_vec, lam)
                b_u[i] = sol[0]
                u[i] = sol[1:]
        for j in range(len(v)): #run ls on v
            X_mat = np.array([np.append(np.array([1]), u[a[0]].T[0]) for a in us_from_v[j]])
            y_vec = np.array([[a[1] - b_u[a[0]]] for a in us_from_v[j]])
            if len(X_mat) > 0:
                sol = solve_ls(X_mat, y_vec, lam)
                b_v[j] = sol[0][0]
                v[j] = sol[:][1:]
            
    # TODO: Evaluate using some error metric measured on test set
    error = 0
    count = 0
    for (a, i, r) in data_test:
        count += 1
        error += (b_u[a] + b_v[i] + np.dot(u[a].T, v[i])[0][0] - r) ** 2
    error = error / count
    return (u, b_u, v, b_v, error)
```

```python
loc = 'ISLR Data'
df = pd.read_csv(loc + '\Advertising.csv')
print(len(df))
df.head()

lambda_list = [10 ** (-.05 * lamb) for lamb in range(-60, 40)] #list of lambda regularizer values
lambda_dic = {lamb: [] for lamb in lambda_list} #going to save the MSE in this dic
iters = 300
for it in range(iters + 1):
    if it%50 == 0:
        print(it) #just to check progress
    for lamb in lambda_list:
        train, test = train_test_split(df, test_size=0.5)
        X_train = np.array(train[['TV', 'Radio', 'Newspaper']])
        X_1 = np.hstack((np.ones(shape = (len(X_train), 1)), X_train))
        y = np.array(train[['Sales']])
        sol = solve_rr(X_1, y, lamb)
        test = np.array(test[['TV', 'Radio', 'Newspaper', 'Sales']])
        MSE = sum([(sol[0] + 
                    sum([test[i][j] * sol[j+1] for j in range(len(sol) - 1)]) -
                    test[i][3]) ** 2 for i in range(len(test))]) / len(test)
        lambda_dic[lamb].append(MSE)
        
mean = np.array([np.mean(lambda_dic[lamb]) for lamb in lambda_list]) #get the mean and std of the MSE for each lambda
std = np.array([1.96 * np.std(lambda_dic[lamb]) for lamb in lambda_list])

plt.figure(1)
p1 = plt.plot(lambda_list, mean)
p2 = plt.fill_between(lambda_list, mean-std, mean+std,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')
plt.legend((p1[0], p2), ('MSE', '95% Confidence Region'))
plt.xscale('log')
plt.xlabel('Lambda')
plt.ylabel('MSE')
plt.title('Ridge Regression Test Error')

plt.show()
```

```python
lambda_list = [10 ** (-.5 * lamb) for lamb in range(-15, 8)]
lambda_dic = {lamb: [] for lamb in lambda_list}
iters = 100
for it in range(iters + 1):
    if it%5 == 0:
        print(it)
    for lamb in lambda_list:
        train, test = train_test_split(df, test_size=0.5)
        X_train = np.array(train[['zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat', 'medv']])
        X_1 = np.hstack((np.ones(shape = (len(X_train), 1)), X_train))
        y = np.array(train[['crim']])
        sol = solve_rr(X_1, y, lamb)
        #crim ~ zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + black + lstat + medv
        test = np.array(test[['zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat', 'medv', 'crim']])
        MSE = sum([(sol[0] + 
                    sum([test[i][j] * sol[j+1] for j in range(len(sol) - 1)]) -
                    test[i][-1]) ** 2 for i in range(len(test))]) / len(test)
        lambda_dic[lamb].append(MSE)
        
mean = np.array([np.mean(lambda_dic[lamb]) for lamb in lambda_list])
std = np.array([1.96 * np.std(lambda_dic[lamb]) for lamb in lambda_list])
```

```python
plt.figure(1)
p1 = plt.plot(lambda_list, mean)
p2 = plt.fill_between(lambda_list, mean-std, mean+std,
    alpha=0.5, edgecolor='#CC4F1B', facecolor='#FF9848')
plt.legend((p1[0], p2), ('MSE', '95% Confidence Region'))
plt.xscale('log')
plt.xlabel('Lambda')
plt.ylabel('MSE')
plt.title('Ridge Regression Test Error')

plt.show()
```
